## 【問題4】学習のためのファイルを作成
新しいデータ（シンプソンズデータセット）を学習します。README.mdのTrainingの項を読み、シンプソンズデータセットを学習するために必要なファイルを作成してください。

アノテーションファイルの形式が問題1の実装とは異なるため変換する必要があります。

```
import re
characters = {'abraham_grampa_simpson': "0",
              'milhouse_van_houten': "1",
              'marge_simpson': "2",
              'chief_wiggum': "3",
              'kent_brockman': "4",
              'sideshow_bob': "5",
              'apu_nahasapeemapetilon': "6",
              'krusty_the_clown': "7",
              'nelson_muntz': "8",
              'edna_krabappel': "9",
              'homer_simpson': "10",
              'ned_flanders': "11",
              'moe_szyslak': "12",
              'principal_skinner': "13",
              'lisa_simpson': "14",
              'bart_simpson': "15",
              'comic_book_guy': "16",
              'charles_montgomery_burns': "17"}

#annotation-file 作成
with open("../ObjectDetection/annotation.txt", "r") as f:
    r = f.read()
    for key, value in characters.items():
      r = r.replace(key+"\n", value+"\n").replace(".jpg,", ".jpg ")
    with open("./train.txt", "w") as w:
      w.write(r)

#classesファイル作成
with open("./classes.txt", "w") as f:
  for character in characters.keys():
    f.write(character+"\n")
```
